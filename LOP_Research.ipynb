```python
import copy
def LOP_ActMat(M,N,k,k0):
    #if k.is_prime()== False:
        #return "Color value must be prime"
    number_of_variables = M*N
    list_of_used = [i for i in range(number_of_variables)]
    square = [list_of_used[i:i+N]  for i in range(0,number_of_variables, N)]
    second_square = [[0 for i in range(number_of_variables)] for j in range(number_of_variables)]
    temp_list = []
    
    row_number = 0
    for i in range (M):
        for j in range (N):
            second_square[row_number][square[i][j]] =1
            if j<N-1:
                second_square[row_number][square[i][j+1]] = 1
            if j !=0:
                second_square[row_number][square[i][j-1]] = 1
            if i <M -1:
                second_square[row_number][square[i+1][j]] =1
            if i != 0:
                second_square[row_number][square[i-1][j]] =1
        
            row_number +=1
    
    #third_square = copy.deepcopy(second_square)
    
    for i in range(number_of_variables):
        second_square[i].append(k0)
    #print(Matrix(third_square)) 
    
    return Matrix(Zmod(k),second_square)
    
#LOP_ActMat(2,2,5,3)
        
```

```python
#Final
import copy
def LOP_Smallest_k0(M,N,k):
    #LOP_All_Solution(num_colors, mat,row_LOP)
    num_colors = k
    mat = LOP_ActMat(M,N,k,0)
    
    second_mat = mat
    matrix = mat
    
    num_rows = matrix.nrows()
    num_cols = matrix.ncols()

    found_solution = False
    entries = 0
    
    reduced_list = []
    for i in range(num_rows):
        for j in range(num_cols-1):
            reduced_list.append(second_mat[i][j])
    reduce_mat = Matrix(num_rows,reduced_list)
        
    first_sol = 0
    
    for i in range(1, int((k+2)/2)):
        if gcd(i,k) < i:
            continue
        else:
            try:
                one_solution = reduce_mat.solve_right(vector([i for j in range(num_rows)]))
                found_solution = True
                return i
            except ValueError:
                found_solution = False

    return first_sol

LOP_Smallest_k0(20,26,8) 
```
